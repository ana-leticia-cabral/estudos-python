2) Faça um programa para calcular e mostrar a maior e menor altura de um grupo de 10 pessoas, a média de altura das mulheres e a quantidade de homens.


```python
def info_10_pessoas():
    lista_nome = []
    lista_altura = []
    lista_sexo = []
    contador = 0

    while len(lista_nome) < 10:
        nome = input(f'Informe o nome da pessoa {contador + 1}: ')
        altura = float(input('Informe a altura (maior que 0): '))
        sexo = input("Informe o sexo (M/F): ").upper()

        nome_valido = all(palavra.isalpha() for palavra in nome.split())
        sexo_valido = sexo in ('M', 'F')

        if nome_valido and altura > 0 and sexo_valido:
            lista_nome.append(nome)
            lista_altura.append(altura)
            lista_sexo.append(sexo)
            contador += 1
        else:
            print("Dados inválidos. Tente novamente.")


    informacoes_pessoas = list(zip(lista_nome, lista_altura, lista_sexo))

    quantidade_mulheres = 0
    altura_mulheres = 0
    quantidade_homens = 0

    for nome, altura, sexo in informacoes_pessoas:
        if sexo == 'F':
            quantidade_mulheres += 1
            altura_mulheres += altura
        elif sexo == 'M':
            quantidade_homens += 1

    maior_altura = max(lista_altura)
    menor_altura = min(lista_altura)
    media_altura_mulheres = altura_mulheres / quantidade_mulheres if quantidade_mulheres > 0 else 0

    for i, (nome, altura, sexo) in enumerate(informacoes_pessoas, 1):
        print(f'Pessoa {i}: {nome}, {altura}m, {sexo}')

    print(f'Maior altura: {maior_altura}\n'
          f'Menor altura: {menor_altura}\n'
          f'Média de altura das mulheres: {media_altura_mulheres:.2f}\n'
          f'Quantidade de homens: {quantidade_homens}')


info_10_pessoas()


```

3) Faça um programa que receba uma letra maiúscula ou minúscula, e mostre todas as letras do alfabeto entre a letra fornecida e a última letra. Exemplo: Entrada = "g"
Saída = "h,i,j,k...z"

```python
def alfabeto():

  alfabeto = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y', 'z']

  letra = input('Digite uma letra: ')

  while letra.lower() not in alfabeto:
    print('Dados inválidos. Digite novamente.')
    letra = input("Digite uma letra: ")


  indice = alfabeto.index(letra.lower())

  resto_alfabeto = [l for l in alfabeto[indice + 1:]]

  print('Restante do alfabeto: ' +', '.join(resto_alfabeto))


alfabeto()
```

5) Faça um programa para calcular o valor de um reajuste em produtos de um mercado. A entrada é composta pelo preço atual do produto e a porcentagem do reajuste. A saída deve conter os novos preços dos produtos. O sistema não pode aceitar valores inválidos de preço.

```python
# Objetivo: criar um algoritmo de reajustes de preços
# Entrada: preço atual e porcentagem do reajuste
# Saída: novos preços dos produtos

produtos = {}
quantidade_produtos = 0

def validar_entradas(valor_atual, porcentagem_reajuste):
  try:
    valor = float(valor_atual)
    porcentagem = float(porcentagem_reajuste)
    return valor, porcentagem
  except ValueError:
    return None

def reajuste(valor_atual, porcentagem_reajuste):

    global quantidade_produtos

    valor, porcentagem = validar_entradas(valor_atual, porcentagem_reajuste)

    if valor is None:
      return 'Entrada inválida!'
    else:
      quantidade_produtos += 1
      novo_valor = valor * (1 + porcentagem / 100)
      produtos[quantidade_produtos] = novo_valor
      return produtos



```

6) Faça um sistema de cadastro de funcionários que tenha como entrada nome do funcionário, nome do departamento, seu salário e o tempo de serviço dele na empresa.
A saída deve ser todos os dados de todos os funcionários e a quantidade de cadastros.
O sistema não deve aceitar valores inválidos para os dados.


```python

# Objetivo: Desenvolver um sistema de cadastro de funcionários que armazene informações
#           relevantes de cada colaborador e mantenha o controle da quantidade de registros.
#
# Entradas:
#   - id          : Identificador único do funcionário (evita conflitos entre nomes iguais)
#   - nome        : Nome completo do funcionário
#   - departamento: Departamento ao qual o funcionário pertence
#   - salario     : Salário atual do funcionário
#   - tempo       : Tempo de serviço na empresa (Vou considerar em meses).
#
# Saídas:
#   - Todos os dados de todos os funcionários
#   - Quantidade total de cadastros realizados
#
# Observação: O campo 'id' foi adicionado para garantir a unicidade de cada registro,
#             já que funcionários podem compartilhar nome, departamento e demais atributos.



# Inicia com um dicionário sem registros e contador em 0.
qtd_cadastros = 0
dados_funcionarios = {}


# Valida as entradas de cada funcionário.
def validar_entradas(id, nome, departamento, salario, tempo):
    try:
        id_func = int(id)
        nome_func = str(nome)
        depart_func = str(departamento)
        salario_func = float(salario)
        tempo_func = int(tempo)
        return id_func, nome_func, depart_func, salario_func, tempo_func
    except ValueError:
        return None


# Para cada funcionário registrado, exibe os dados de todos os funcionários.
def exibir_funcionarios():
    print('Informações dos funcionários cadastrados:')
    print('------------------------------------')
    for id_key, dados in dados_funcionarios.items():
        print(f'  Id: {id_key}')
        for chave, valor in dados.items():
            print(f'  {chave}: {valor}')
        print('------------------------------------')
    print(f'Total de cadastros: {qtd_cadastros}\n')


# Função que armazena os registros
def cadastro_funcionarios(id, nome, departamento, salario, tempo):
    validacao = validar_entradas(id, nome, departamento, salario, tempo)

    if validacao is not None:
        id, nome, departamento, salario, tempo = validacao

        if nome.replace(" ", "").isalpha() and departamento.replace(" ", "").isalpha():
            dados_funcionarios[id] = {'Nome': nome, 'Departamento': departamento, 'Salário': salario, 'Tempo de serviço': tempo}
            global qtd_cadastros
            qtd_cadastros += 1
        else:
            return 'Dados inválidos!'
    else:
        return 'Dados inválidos!'

    return exibir_funcionarios()

cadastro_funcionarios(1, 'Aline Silva', 'Dados', 1000, 9)
cadastro_funcionarios(4, 'Carlos Silva', 'TI', 2000, 24)
```

7) Faça um código que decodifique a língua do p, e mostre a frase em língua portuguesa. A entrada é uma frase na língua do p, a saída é uma frase normal.


```python
# Objetivo: Desenvolver um código que decodifique a língua do p.
#
# Entradas:
#   - frase_p: frase na língua do p
#
# Saídas:
#   - frase_normal

def decodifica_lingua_p(frase_p):

  frase_p = frase_p.lower()

  if 'p' in frase_p:
    frase_normal = frase_p.replace('p', '')
    return frase_normal


decodifica_lingua_p('PAmanhã pé pum pnovo pdia!')
```

8. Faça um jogo de par ou ímpar, onde o jogo recebe os nomes dos jogadores e ao final da execução mostra o resultado de todas as rodadas e o vencedor.
O primeiro jogador sempre escolhe ímpar e o segundo sempre escolhe par.


```python
# Objetivo:
#   - Desenvolver um código que crie um jogo com o total de 2 jogadores
#   - O primeiro jogador deverá SEMPRE escolher um número ímpar
#   - O segundo  jogador deverá SEMPRE escolher um número par
#   - OS dois números escolhidos deverão ser somados
#   - Se a soma resultar em um número ímpar, jogador 1 vence. Caso contrário, jogador 2 vence.
#
# Entradas:
#   - jogador_1: nome do primeiro jogador
#   - jogador_2: nome do segundo jogador
#   - numero_1: número do primeiro jogador
#   - numero_2: número do segundo jogador
#
# Saídas:
#   - Imprimir na tela o resultado de todas as rodadas e o vencedor



def jogo_par_impar():

    jogador_1 = input('Digite o nome do primeiro jogador: ')
    jogador_2 = input('Digite o nome do segundo jogador: ')

    if jogador_1.replace(" ", "").isalpha() and jogador_2.replace(" ", "").isalpha():
        continuar = True

        while continuar:

            try:
                numero_1 = int(input(f'{jogador_1}, escolha um número ÍMPAR: '))
            except ValueError:
                print('Digite um número inteiro!')
                continue

            if numero_1 % 2 == 0:
                print(f'{jogador_1}, você deveria ter escolhido um número ÍMPAR. '
                      f'Iniciem o jogo novamente e insiram os dados corretos!')
                continuar = False

            else:
                try:
                    numero_2 = int(input(f'{jogador_2}, escolha um número PAR: '))
                except ValueError:
                    print('Digite um número inteiro!')
                    continue

                if numero_2 % 2 != 0:
                    print(f'{jogador_2}, você deveria ter escolhido um número PAR. '
                          f'Iniciem o jogo novamente e insiram os dados corretos!')
                    continuar = False

                else:
                    soma = numero_1 + numero_2
                    print('--------------------')

                    if soma % 2 == 0:
                        print(f'A soma dos dois números escolhidos é: {soma}, sendo um número PAR.\n'
                              f'Ganhador do jogo: {jogador_2}\n')
                    else:
                        print(f'A soma dos dois números escolhidos é: {soma}, sendo um número ÍMPAR.\n'
                              f'Ganhador do jogo: {jogador_1}\n')

                    continuar_jogo = input('Desejam continuar o jogo? S (sim) ou N (não): ')
                    if continuar_jogo.upper() == 'N':
                        continuar = False
    else:
        print('Insira nomes válidos!')
```

9. Faça um programa que receba vários números e mostre ao final o segundo e terceiro menor número


```python
def numeros():
    numeros = []
    continuar = True

    while continuar:
        try:
            n = int(input('Digite os números desejados (mínimo 3): '))
        except ValueError:
            print('Por favor, digite um número inteiro!')
            continue

        numeros.append(n)

        escolha = input('Deseja continuar? S (sim) ou N (não)? ')

        if escolha.upper() == 'N':
            if len(numeros) >= 3:
                continuar = False
            else:
                print(f'É necessário informar ao menos {3 - len(numeros)} número(s) a mais.')
        elif escolha.upper() == 'S':
            print('Vamos continuar...')
        else:
            print('Considerando que queira continuar...')

    lista_ordenada = sorted(numeros)

    print(f'Lista de números fornecida: {numeros}')
    print(f'Segundo menor número da lista: {lista_ordenada[1]}')
    print(f'Terceiro menor número da lista: {lista_ordenada[2]}')

numeros()
```

10. Faça um programa que receba vários números positivos, e mostre as posições dos números pares e dos números ímpares.

```python
def numeros_positivos():
    numeros_positivos = []
    continuar = True

    while continuar:
        try:
            n = int(input('Digite os números desejados (maior que 0): '))
        except ValueError:
            print('Por favor, digite um número inteiro!')
            continue

        if n <= 0:
            print('É necessário informar um número positivo!')
            continue

        numeros_positivos.append(n)

        escolha = input('Deseja continuar? S (sim) ou N (não)? ')

        if escolha.upper() == 'N':
            continuar = False
        elif escolha.upper() == 'S':
            print('Vamos continuar...')
        else:
            print('Considerando que queira continuar...')

    pares = [i for i, n in enumerate(numeros_positivos) if n % 2 == 0]
    impares = [i for i, n in enumerate(numeros_positivos) if n % 2 != 0]
    print(f'Lista de números fornecida: {numeros_positivos}')
    print(f'Posições dos números pares: {pares}')
    print(f'Posições dos números ímpares: {impares}')

numeros_positivos()
```

11. Faça um programa para um concurso de piadas, onde a entrada é o número de juízes e as notas de cada juiz.
A primeira linha contém as notas do primeiro juiz, a segunda linha contém a nota do segundo juiz e assim por diante. A saída é o resultado do concurso, mostrando o número do ganhador.


```python
import pandas as pd

# Objetivo:
#   - Desenvolver um código para um concurso de piadas
#   - A primeira linha deverá conter as notas do primeiro juiz
#   - A segunda linha deverá conter as notas do terceiro juiz ...
#
# Entradas:
#   - qtd_juizes: número de juízes
#   - notas: nota de cada juiz para cada participante
#
# Saídas:
#   - Resultado do concurso e o número do ganhador

def concurso(qtd_juizes, notas):

  juizes = range(1, qtd_juizes + 1)

  participante_1 = [nota for juiz, participantes in notas.items() for participante, nota in participantes.items() if participante == "Participante 1"]

  participante_2 = [nota for juiz, participantes in notas.items() for participante, nota in participantes.items() if participante == "Participante 2"]

  participante_3 = [nota for juiz, participantes in notas.items() for participante, nota in participantes.items() if participante == "Participante 3"]


  df = pd.DataFrame({
    "Juiz": juizes,
    "Participante 1": participante_1,
    "Participante 2": participante_2,
    "Participante 3": participante_3
  })

  print(f'Resultado do concurso: \n'
        f'{df.to_string(index = False)}\n')
  print(f'Ganhador do concurso: {df.drop(columns="Juiz").sum().idxmax()}')




notas = {
    "Juiz 1": {"Participante 1": 8.5, "Participante 2": 7.0, "Participante 3": 9.2},
    "Juiz 2": {"Participante 1": 9.0, "Participante 2": 8.3, "Participante 3": 7.5},
    "Juiz 3": {"Participante 1": 7.8, "Participante 2": 9.1, "Participante 3": 8.0},
    "Juiz 4": {"Participante 1": 9.3, "Participante 2": 8.7, "Participante 3": 7.2},
    "Juiz 5": {"Participante 1": 8.1, "Participante 2": 7.5, "Participante 3": 9.4},
}

concurso(5, notas)
```
